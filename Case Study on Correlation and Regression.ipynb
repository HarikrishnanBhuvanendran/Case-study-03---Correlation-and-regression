A textile company regularly spent money on five advertising companies for their ads.

1) First of all, check whether there is any relationship between the amount spent for advertising through five different companies and profit generated.

2) Build a regression model to predict the profit from the given data and comment your findings on it.

```python
#import all the relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset 
data= pd.read_csv("Add_sales_data_2018.csv")
```

```python
#checking head of the dataset
data.head()
```

```python
#checking info of the dataset
data.info()
```

```python
#checking is there any null values in the dataset
data.isna().sum()
```

# 1. First of all, check whether there is any relationship between the amount spent for advertising through five different companies and profit generated

```python
# we have to find out,is there any relationship between the amount spent for advertising through five different companies and profit generated
corrdata=data.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrdata,annot=True,linewidth=0.1,cmap="terrain")
```

# Insights
>From the above heatmap it's clear that ,the amount spent for advertising through five different companies and profit generated are stongly correlated.

# 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
#checking the columns
data.columns
```

```python
#predict the profit from the given data
#put the profit(in thousands)in y variable
y=data["Profit(in Thousands)"]
#drop the unwanted columns in x variable
x=data.drop(['Profit(in Thousands)','Date'],axis=1)
```

```python
#import scikitlearn libraries for train_test_split
#splitting the x & y variables into training and testing sets
from sklearn. model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
#finding linear regression
from sklearn import linear_model
lr= linear_model.LinearRegression()
```

```python
#Here we're creating a model
model= lr.fit(x_train,y_train)
```

```python
#find predictions
predictions=model.predict(x_test)
```

```python
#find the y_test
print(y_test)
```

```python
print(predictions)
```

```python
#import mean_squared_error from scikitlearn for finding MSE
from sklearn.metrics import mean_squared_error
```

```python
#checking the efficiency of the model by finding mean squared error value.
print("MSE: is",mean_squared_error(y_test,predictions))
#Here we add some codes to find out the rounded values.
print("Rounded  is MSE : %0.2f "%mean_squared_error(y_test,predictions))
```

# Insights : 
    The mean squared error is approximately equal to zero ,
    that means the predicted values and actual values are so close .
    Hence the model can be used for predictions
